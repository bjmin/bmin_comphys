## Legendre Function

```python
from sympy import *
%matplotlib inline
init_printing()
```

```python
x, t = symbols("x,t")
# n-th order Legendre Polynomial legendre(n,x)
legendre(0,x), legendre(1,x), legendre(2,x), legendre(3,x)
```

```python
plot(legendre(0,x),legendre(1,x),legendre(2,x),legendre(3,x),(x,-1,1))
```

## Bessel Function

```python
# besselj(n,x): Bessel function 1st kind
# bessely(n,x): Bessel function 2st kind
plot(besselj(0,x),besselj(1,x),besselj(2,x),(x,0,10))
```

```python
plot(bessely(0,x),bessely(1,x),bessely(2,x),(x,0,10),ylim=(-1,1))
```

## Gamma Function

```python
# gamma(n)
gamma(1), gamma(2), gamma(3), gamma(4)
```

```python
plot(gamma(x),(x,-5,5),ylim=(-10,10))
```
