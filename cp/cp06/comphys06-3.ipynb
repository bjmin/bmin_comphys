# 카오스

비선형 미분방정식의 수치적 풀이를 통해 흥미로운 현상을 발견할 수 있습니다. 다음처럼 주어지는 로렌츠 방정식을 살펴봅시다.  
$$ \frac{dx}{dt} = s (y-x),\\
\frac{dy}{dt} = x(r -z) -y, \\
\frac{dz}{dt} = xy - bz $$
이 방정식은 1963년 에드워드 로렌츠가 결정론적 카오스를 발견할 때 사용한 식입니다.  
* 로렌츠 시스템 : https://en.wikipedia.org/wiki/Lorenz_system

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

r = 28.0
s = 10.0
b = 8.0 / 3.0

def f(state, t):
    x, y, z = state  
    return s * (y - x), x * (r - z) - y, x * y - b * z 

state0 = [0.0, 1.0, 0.0]
t = np.arange(0.0, 100.0, 0.01)
states = odeint(f, state0, t)
plt.plot(t,states[:,0])
```

```python
# z vs. x를 그리면 다음과 같은 strange attractor를 관찰할 수 있다.
plt.plot(states[:,2],states[:,0])
```

```python
# 조건을 조금 바꾸어보면 매우 다른 동역학을 관찰할 수 있다. 
r, s, b = 14.0, 10.0, 8.0/3.0
state0 = [0.0, 1.0, 0.0]
states = odeint(f, state0, t)
plt.plot(states[:,2],states[:,0])
```
