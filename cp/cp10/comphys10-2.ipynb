# Gradient Descent

fitting 문제는 오차를 최소가 되는 지점을 찾는 최적화(optimization) 문제의 일부분입니다. 간단한 최적화 문제를 살펴봅시다.

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(0,10,100)
def f(x):
    return 4*(x-4)+(x-5)**2
plt.plot(z,f(z))
```

다음과 같은 모양의 퍼텐셜에서 최소값을 찾으려면 임의의 지점에서 시작해서 퍼텐셜이 낮아지는 방향으로 조금씩 움직이면 된다. 즉 특정 지점 $x$에서 기울기가 양수이면 $x$를 감소시키고, 기울기가 음수이면 $x$를 증가시키면서 최소값을 찾으면 된다. 이를 경사하강법이라고 한다. 

구체적으로 다음의 과정을 반복하면 된다.
$$ x \rightarrow x - \gamma f'(x).$$
여기에서 $f'(x)$는 기울기를 나타내며 $\gamma$는 $x$를 얼마나 빨리 바꿀지를 결정하는 상수이다.

$\gamma$가 너무 크면 최소값을 빨리 찾지 못하고 진동할 것이고 너무 작으면 최소값을 찾을 때까지 지나치게 많은 시간이 걸릴 것이다. 또한 $\gamma$가 음수가 되면 최대값을 찾을 것이다.

위의 예시에서 최소값을 찾아가는 과정을 살펴보자.

```python
def gd(gamma,x):
    dx = 0.001
    df = (f(x+dx)-f(x))/dx # 기울기
    return x-gamma*df
```

```python
x0=10
x = [x0]
y = [f(x0)]
gam=0.2

for i in range(10):
    x0=gd(gam,x0)
    x=np.append(x,x0)
    y=np.append(y,f(x0))

plt.plot(z,f(z))
plt.plot(x,y,'o')
```

만일 퍼텐셜이 2차원 이상일 경우에도 같은 방식으로 계산해주면 된다. 단 다변수이므로 기울기가 Gradient가 된다. 즉, 다음의 과정을 반복하면 된다.

$$ x \rightarrow x - \gamma \nabla f(x).$$
