## simplify & factor & expand 
* 식 정리하기

```python
x, y, z = sympy.symbols("x,y,z")

f1=2*sympy.cos(x)*sympy.sin(x)
f1
```

```python
sympy.simplify(f1)
```

```python
f2 = 2*(x**2-x)-x*(x+1)
f2
```

```python
sympy.simplify(sympy.exp(x)*sympy.exp(3*x))
```

```python
f3= x**3 + 3*x**2*y + 3*x*y**2 + y**3
sympy.factor(f3)
```

```python
sympy.factor( x**2+7*x+12)
```

```python
sympy.expand((x+y)**3)
```

## solve

```python
sympy.solve(x**2+2*x-3)
```

## diff & integrate
* 미분 적분

$$ x e^{-x^2} $$

```python
x,y,z = sympy.symbols("x,y,z")

sympy.diff(x* sympy.exp(-x**2), x)
```

```python
sympy.diff(sympy.sin(x),x), sympy.diff(sympy.sin(x),x,2)
```

```python
sympy.integrate(sympy.sin(x),x)
```

```python
sympy.integrate(sympy.sin(x)+sympy.cos(y),x)
```

```python
sympy.integrate(sympy.exp(-x**2),x)
```

```python
sympy.integrate(sympy.exp(-x),(x,a,b)) 
```

```python
sympy.integrate(sympy.exp(-x**2),(x,-sympy.oo,sympy.oo)) # double o (oh) represents infinity 
```
