# 3. sinc function

1. 싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.
$$ sinc(x)  = \frac{\sin x}{x}. $$
2. 싱크 함수를 테일러 전개하면 다음과 같다.
$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$ 
위의 식의 우변을 $n$을 바꾸어 가면서  그래프로 그리고 싱크함수와 비교하여라.
