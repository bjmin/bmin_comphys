# 함수(function)

```python
from math import *
import numpy as np

f = 1.0
for i in range(1,10):
    f *= i
f
```

```python
def facto(x):
    f = 1.0
    for i in range(1,x+1):
        f *= i
    return f
facto(10)
```

## 예제 1 : 자유낙하 운동 거리 계산

```python
# 시간 t동안 자유낙하하는 거리 계산 
def falling(g,t):
    g=9.8
    x=0.5*g*t**2
    return x

falling(9.8,2),falling(9.8,4)
```

```python
# 높이 h에서 떨어지는 물체의 시간 t에서의 수직 위치
def fall_y(g,h,t):
    g=9.8
    x=h-0.5*g*t**2
    if(x<0):
        x=0
    return x

fall_y(9.8,30,2), fall_y(9.8,30,3)
```

## 예제 2 : 포물선 운동
* 초기속도 (v_x,v_y)
* 가속도 (0,-g)

```python
g=9.8
def para(v,t):
    x=[v[0]*t,v[1]*t-0.5*g*t**2]
    return x

v0=[1,2]
para([1,2],2), para(v0,2), para(np.array([1,2],float),2)
```
