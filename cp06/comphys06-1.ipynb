# 행렬(matrix)
* 하나의 1차방정식 : 초등, 중학교 수학교과서를 참조하여 스스로 푼다.
* 연립방정식 : 가우스 소거법(Gaussian elimination)
* LU decomposition: numpy.linalg.solve

## Array! Hooray!

```python
import numpy as np
M0 = np.array([[1,0],[0,1]])
M1 = np.array([[1,2],[3,4]])
```

```python
print(3*M1, M0+M1, M0*M1)
```

```python
np.dot(M1,M1)
```

```python
np.transpose(M1), np.linalg.inv(M1), np.linalg.eig(M1), np.linalg.det(M1)
```
