```python
x = np.arange(0, 2*np.pi, 0.5)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.axis([0,2*np.pi,-2,2])
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x)')
plt.plot(x, y1, color='green',marker='o',linestyle='--',label='sin')
plt.plot(x, y2, color='red',marker='x',linestyle='-',label='cos');
plt.plot(x, y3, color='blue',marker='s',linestyle='-.',linewidth=4,label='tan');
plt.legend()
#plt.savefig('./tri_figure.pdf')
```

```python
import csv

with open('./data/countries.csv', newline='') as f:
    reader = csv.reader(f)
    countries = list(reader)

name=[str(x[1]) for x in countries[1:]]
# n_italy=[[float(x[2]),float(x[3]),float(x[4])] for x in countries[1:] if x[1]=='Italy']
```
